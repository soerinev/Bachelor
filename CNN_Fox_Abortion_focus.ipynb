```python
# read csv
file_path = "/work/Bachelor/sample_data/dropbox/joined_dataframe/joined_17_05.csv"
joined_17_05 = pd.read_csv(file_path)
joined_17_05
```

```python
# filter articles to only have articles from CNN and Fox News
CNN_articles = joined_17_05[joined_17_05['source'] == 'CNN']
Fox_articles = joined_17_05[joined_17_05['source'] == 'Fox News'] #ðŸ‘º there are no fox news articles in the sample, so I'm using 'New York Daily News' as a proxy. Be sure to change in real analysis!!

# combine dfs into a new one
CNN_Fox_articles = pd.concat([CNN_articles, Fox_articles])

# display the new df containing only CNN and Fox News articles
CNN_Fox_articles
#len(CNN_Fox_articles)
```

```python
# save csv
output_file_path = "/work/Bachelor/sample_data/dropbox/joined_dataframe/CNN_Fox_joined_17_05.csv"
CNN_Fox_articles.to_csv(output_file_path, index=False, encoding='utf-8')
```

### How to only include articles related to abortion

```python
# Search for rows where the 'text' column contains the word 'abortion'
abortion_articles = CNN_Fox_articles[CNN_Fox_articles['body'].str.contains('abortion', case=False, na=False)]
abortion_articles
#len(abortion_articles) # there's 11 here
```

#### Wider search

```python
# Search for articles with any of the abortion related terms
search_terms = r"abortion|reproductive rights|pro-choice|pro-life|Roe v\. Wade|Dobbs v\. Jackson Women's Health Organization|planned parenthood|family planning"

# Filter articles that contain any of the search terms
abortion_articled_joined_17_05 = CNN_Fox_articles[CNN_Fox_articles['body'].str.contains(search_terms, case=False, na=False, regex=True)] #ðŸ‘º change to the CNN_Fox_articles df. I used text_sources_df because there's no abortion articles among the CNN_Fox_articles

# Display the filtered articles
abortion_articled_joined_17_05
#len(abortion_articles) # there's 14 here compared to 11 when you just search for 'abortion'
```

```python
# save csv
output_file_path = "/work/Bachelor/sample_data/dropbox/joined_dataframe/abortion_articled_joined_17_05.csv"
abortion_articled_joined_17_05.to_csv(output_file_path, index=False, encoding='utf-8')
```
